```python
#Se usar o arquivo descompactado
dados = pd.read_csv('DOM2015.csv',sep=',')

#Se usar o arquivo compactado como está no dropbox púbico
#pd.read_csv('DOM2015csv.zip', compression='zip', header=0, sep=',', quotechar='"')
```

```python
dados.loc[dados.V4105<=3, 'RU'] = 'Urbano'
dados.loc[dados.V4105>3, 'RU'] = 'Rural'
```

```python
#dados.V0227.dtypes
```

```python
dados.loc[(dados.UF<20),'Regiao'] = 'Norte'
dados.loc[(dados.UF>=20),'Regiao'] = 'Nordeste' #| dados.UF<30
dados.loc[(dados.UF>=30),'Regiao'] = 'Sudeste' #| dados.UF<40
dados.loc[(dados.UF>=40), 'Regiao'] = 'Sul' #| dados.UF<50
dados.loc[(dados.UF>=50), 'Regiao'] = 'Centro-Oeste' #| dados.UF<60
```

```python
#Limitações
dados = dados[dados.RU == 'Urbano'] #Area Urbana
dados = dados[dados.V0203 <= 2]   #Sem palha
dados = dados[dados.V0219 == 1] #Energia Elétrica
```

```python
#Número de pessoas
dados.loc[(dados.V0106 == float(1)), 'Npessoas'] = '1 pessoa'
dados.loc[(dados.V0106 == float(2)), 'Npessoas'] = '2 pessoas'
dados.loc[dados.V0106 == 3, 'Npessoas'] = '3 pessoas'
dados.loc[dados.V0106 == 4, 'Npessoas'] = '4 pessoas'
dados.loc[dados.V0106 == 5, 'Npessoas'] = '5 pessoas'
dados.loc[dados.V0106 >= 6, 'Npessoas'] = '6 ou mais pessoas'
dados.Npessoas.fillna('Não se aplica')
dados.Npessoas.value_counts(True)*100
```

```python
#Tipo de acesso
dados.loc[dados.V02424==2,'TipoIN'] = '3G ou 4G'
dados.loc[dados.V02425==2,'TipoIN'] = 'Conexão discada'
dados.loc[dados.V02426==2,'TipoIN'] = 'Banda larga'
dados.TipoIN.fillna('Não se aplica')
dados.TipoIN.value_counts(True)*100
```

```python
#Aparelho de acesso a internet
dados.loc[dados.V02323==2,'Aparelho'] = 'Microcomputador'
dados.loc[dados.V02324==2,'Aparelho'] = 'Telefone celular'
dados.loc[dados.V02325==2,'Aparelho'] = 'Tablet'
dados.loc[dados.V02326==2,'Aparelho'] = 'TV'
dados.loc[dados.V02327==2,'Aparelho'] = 'Outro'

dados.Aparelho.fillna('Não se aplica')
dados.Aparelho.value_counts(True)*100
```

```python
#Renda per capita
dados.loc[dados.V4622== 0,'Renda'] = 'Sem rendimento'
dados.loc[dados.V4622== 1,'Renda'] = 'Até 1/4 salários mínimos'
dados.loc[dados.V4622==2,'Renda'] = '1/4 - 1/2 salários mínimos'
dados.loc[dados.V4622==3,'Renda'] = '1/2 - 1 salários mínimos'
dados.loc[dados.V4622==4,'Renda'] = '1 - 2 salários mínimos'
dados.loc[dados.V4622==5,'Renda'] = '2 - 3 salários mínimos'
dados.loc[dados.V4622==6,'Renda'] = '3 - 5 salários mínimos'
dados.loc[dados.V4622==7,'Renda'] = ' Mais de 5 salários mínimos'
dados.Renda.fillna('Não se aplica')
dados.Renda.value_counts(True)*100
```

```python
#Agua encanada
dados.loc[dados.V4624<= 3,'Agua'] = 'Canalização interna'
dados.loc[dados.V4624> 3,'Agua'] = 'Canalização externa'
dados.Agua.fillna('Não se aplica')
dados.Agua.value_counts(True)*100
```

```python
#Tem TV
dados.loc[(dados.V0226==float(2))| (dados.V0227==float(1)), 'TV'] = 'Sim'
dados.loc[(dados.V0226==float(4))| (dados.V0227==float(3)), 'TV'] = 'Não'
dados.TV.fillna('Não se aplica')
dados.TV.value_counts(True)*100
```

```python
#Cores da TV
dados.loc[(dados.V0226==float(2)), 'CorTV'] = 'Cores'
dados.loc[(dados.V0227==float(1)), 'CorTV'] = 'Preto e Branco'
dados.CorTV.fillna('Não se aplica')
dados.CorTV.value_counts(True)*100
```

```python
#Tipo de TV
dados.loc[(dados.V02272==float(1)), 'TipoTV'] = 'Assinatura'
dados.loc[(dados.V02273==float(1)), 'TipoTV'] = 'Sinal digital aberto'
dados.loc[(dados.V02274==float(1)), 'TipoTV'] = 'Antena Parabólica'
dados.TipoTV.fillna('Não se aplica')
dados.TipoTV.value_counts(True)*100
```

```python
r = dados.columns.get_loc('Regiao')
t = dados.columns.get_loc('TipoTV')
dados = dados.iloc[:,r:t]
dados.fillna('Não se aplica')
```
